```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, Eq, diff
import  random
x = symbols('x')

''' X_{n+1} = X_{n} - \frac{f(X_n)}{f'(X_n)}'''
def calc_error(punto_n, punto_v):
    e = abs((punto_n - punto_v) / punto_n)
    return e
    
def newton_raphson(funcion, error_g):
    g_func = eval(funcion)
    g_func_prima = g_func.diff(x)
    error = 1.0
    iteracion = 0
    X_val = random.uniform(1, 100)  
    X_new = 0
    while error >= error_g:
        X_new = X_val - ((g_func.evalf(subs={x: X_val}))/(g_func_prima.evalf(subs={x: X_val})))
        error = calc_error(X_new,X_val)
        iteracion +=1
        print(f'Iteración {iteracion}: {X_new}')
        X_val = X_new 
    return iteracion, X_new

iter, root = newton_raphson("x**x -100",10*10**(-5))
print(iter,root)
```

```python
from sympy import symbols

x = symbols('x')

def bisection_method(funcion, error):
    funcion = eval(funcion)
    a = -100.0
    b = 100.0
    iterations = 0
    y_value = 0

    while (b - a) / 2 >= error:
        c = (b + a) / 2.0
        y_value = c
        if funcion.evalf(subs={x: c}) >= 0:
            b = c
        else:
            a = c
        iterations += 1
        

    return iterations, y_value

iteraciones, root = bisection_method("x**x-100", 10 * 10 ** (-4))
print(f'Número de iteraciones: {iteraciones}')
print(f'Raíz aproximada: {root}')
```
